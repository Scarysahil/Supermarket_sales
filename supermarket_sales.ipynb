```python
import matplotlib.pyplot as plt

sale = df.groupby('Branch')['Sales'].sum()

plt.figure(figsize=(8, 6))  # set a sensible figure size
sale.plot(kind='bar')

for i, val in enumerate(sale):
    plt.text(i, val + 10, str(val), ha='center')

plt.ylabel("Number of sales")
plt.title("Total sales by branch")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns 
month = df.groupby("Month")['Sales'].sum()
explode = [0, 0.1, 0.1]  
plt.pie(month.values, labels=month.index, autopct="%1.1f%%", shadow=True, explode=explode, wedgeprops= {'edgecolor': 'black'})

```
